```python
import pandas as pd
from epftoolbox.data import read_data
```

**read_data()**

If `dataset` is either ``"PJM"``, ``"NP"``, ``"BE"``, ``"FR"``, or ``"DE"``,
    the function checks whether ``dataset.csv`` exists in ``path``. If it doesn't exist,
    it downloads the data from an online database and saves it under the ``path`` directory. ``"PJM"``

```python
# '.' means current directoory
df_train, df_test = read_data(path='.', dataset='PJM', begin_test_date='01-01-2016', end_test_date='01-02-2016')

print(df_train.shape,df_test.shape)

df_train.head()
# df_train.tail()
```

```python
# df_test.head()
df_test.tail()
```

```python
# for val in df_test.values:
  # print(val)
```

**Data Scaling**

```python
from epftoolbox.data import DataScaler

# Training Data
Xtrain = df_train.values

scaler = DataScaler('Norm')
Xtrain_scaled = scaler.fit_transform(Xtrain)
Xtrain_inverse = scaler.inverse_transform(Xtrain_scaled)

print(Xtrain.shape,Xtrain_scaled.shape,Xtrain_inverse.shape)

print("\n Sample Xtrain \n\n",Xtrain[:3:])
print("\n Sample Xtrain_scaled \n\n",Xtrain_scaled[:3:])
print("\n Sample Xtrain_inverse\n\n",Xtrain_inverse[:3:])
```

```python
# Test Data
Xtest  = df_test.values
Xtest_scaled  = scaler.transform(Xtest)
Xtest_inverse = scaler.inverse_transform(Xtest_scaled)

print(Xtest.shape,Xtest_scaled.shape,Xtest_inverse.shape)

print("\n Sample Xtest \n\n",Xtest[:3:])
print("\n Ssample Xtest_scaled \n\n",Xtest_scaled[:3:])
print("\n Sample Xtest_inverse \n\n",Xtest_inverse[:3:])
```

**Forecasting Models**

```python
from epftoolbox.evaluation import MAPE
```

```python
path = '/content/epftoolbox/forecasts/Forecasts_NP_DNN_LEAR_ensembles.csv'

forecast = pd.read_csv(path,index_col = 0)

print(forecast.shape)
forecast.head()
```

```python
forecast.columns
```

```python
# Transforming indices to datetime format

forecast.index = pd.to_datetime(forecast.index)
print(forecast.index[:3])
```

```python
_, df_test = read_data(path='.', dataset = 'NP', begin_test_date = forecast.index[0],end_test_date = forecast.index[-1])

# we are only interested in test data as the forecast is done in test data itself.
print(df_test.shape)
print(df_test.head(3))
print(df_test.tail(2))
```

```python
# Extracting forecast of DNN ensemble and display
fc_DNN_ensemble = forecast.loc[:, ['DNN Ensemble']]

fc_DNN_ensemble[:3]
```

```python
# Extracting real price and display
real_price = df_test.loc[:, ['Price']]

real_price[:25]
```

```python
print(fc_DNN_ensemble.shape)
fc_DNN_ensemble.values.reshape(-1, 24).shape

# making data according to the 24 our time span thus if our dataaset size is n then 
# number of cols will be 24 and number of rows will be -1 that is number of rows = n//24
# print(fc_DNN_ensemble.shape[0]//24)
```

```python
fc_DNN_ensemble_2D = pd.DataFrame(fc_DNN_ensemble.values.reshape(-1, 24), index = fc_DNN_ensemble.index[::24], 
                                  columns=['h' + str(hour) for hour in range(24)])

fc_DNN_ensemble_2D.head()
```

```python
real_price_2D = pd.DataFrame(real_price.values.reshape(-1, 24), index=real_price.index[::24], 
                             columns=['h' + str(hour) for hour in range(24)] )

real_price_2D.head()
```

* According to the paper, the MAPE of the DNN ensemble for the NP market is 5.38%.

```python
# Evaluating MAPE when real price and forecasts are both dataframes

MAPE(p_pred = fc_DNN_ensemble, p_real = real_price) * 100
```

```python
# Evaluating MAPE when real price and forecasts are both numpy arrays

MAPE(p_pred = fc_DNN_ensemble.values, p_real = real_price.values) * 100
```

```python
# Pandas Series

MAPE(p_pred=fc_DNN_ensemble.loc[:, 'DNN Ensemble'], p_real=real_price.loc[:, 'Price']) * 100
```

```python
# Numpy arrays

MAPE(p_pred=fc_DNN_ensemble.values.squeeze(), p_real=real_price.values.squeeze()) * 100
```

```python
# Evaluating MAPE when input values are of shape (ndays, n_prices/day) instead 
# of shape (nprices, 1)
# Dataframes

MAPE(p_pred=fc_DNN_ensemble_2D, p_real=real_price_2D) * 100
```
